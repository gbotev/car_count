```python
'''
Other functions'''
#from https://www.pyimagesearch.com/2017/01/02/rotate-images-correctly-with-opencv-and-python/
def rotate_bound(image, angle):
    # grab the dimensions of the image and then determine the
    # center
    (h, w) = image.shape[:2]
    (cX, cY) = (w // 2, h // 2)
 
    # grab the rotation matrix (applying the negative of the
    # angle to rotate clockwise), then grab the sine and cosine
    # (i.e., the rotation components of the matrix)
    M = cv2.getRotationMatrix2D((cX, cY), -angle, 1.0)
    cos = np.abs(M[0, 0])
    sin = np.abs(M[0, 1])
 
    # compute the new bounding dimensions of the image
    nW = int((h * sin) + (w * cos))
    nH = int((h * cos) + (w * sin))
 
    # adjust the rotation matrix to take into account translation
    M[0, 2] += (nW / 2) - cX
    M[1, 2] += (nH / 2) - cY
 
    # perform the actual rotation and return the image
    return cv2.warpAffine(image, M, (nW, nH))

#from https://github.com/Guanghan/darknet/blob/master/scripts/convert.py
#down in the notebook

def check_data_folders(dir_name):
    if not os.path.isdir(dir_name):
        os.mkdir(dir_name)
        for f in ['images', 'labels', 'labels_not_yolo']:
            os.mkdir(os.path.join(dir_name, f))
```

### Workflow:
1. go through folders
2. go through images and open both real and annotated ones
3. for N steps perform random permutations to BOTH images
    * for each step get the labels in the form needed for YOLO, write them in a file and write-out the image to a folder.


```python
class satellite_image:
    '''
    This class is for generating patches from a large satellite image
    '''
    
    #min and max sides of the patch
    __min_side = 96
    __max_side = 256
    
    def __init__(self, im_name):
        self.im = cv2.imread(im_name)
        self.im_annotated_cars = cv2.imread(im_name[:-4] + "_Annotated_Cars.png")
        
    def set_min_side(self, min_side):
        self.__min_side = min_side
        
    def set_max_side(self, max_side):
        self.__max_side = max_side
        
    def show_image(self):
        plt.imshow(self.im)
        
    def get_random_car_patch(self, rotate=True):
        patch = np.zeros((0,0,0))
        while len(patch) == 0:
            (x, y, side) = self.__create_random_x_y()      
            rot_angle = np.random.randint(359)
            annotated_patch = self.im_annotated_cars[y:y+side, x:x+side, 2]
            if rotate:
                annotated_patch = rotate_bound(annotated_patch, rot_angle)                
            car_coords = annotated_patch.nonzero()
            #if there are cars annotated in the patch
            if len(car_coords[0]) > 0:
                if rotate:
                    #when we rotate a pixel from the first image, transforms into 1 to 4 pixels in the new one
                    #also they can range in +/-2 pixels 
                    coords_0 = [car_coords[0][0]]
                    coords_1 = [car_coords[1][0]] 
                    for i in range(1, len(car_coords[0])):
                        if (abs(car_coords[0][i] - car_coords[0][i-1]) > 4 
                                or abs(car_coords[1][i] - car_coords[1][i-1]) > 4):
                            coords_0.append(car_coords[0][i])
                            coords_1.append(car_coords[1][i])
                    car_coords = np.array([coords_0, coords_1])
                    (patch, coords) = self.__generate_car_coords(car_coords, x, y, side, rot_angle)
                else:
                    (patch, coords) = self.__generate_car_coords(car_coords, x, y, side)
        return(patch, coords)
        
    
    def __create_random_x_y(self):
        #select a patch side within 96 (2 cars) to 256 (6 cars) side
        side = self.__min_side + np.random.randint(self.__max_side - self.__min_side)
        #select random x, y starting point
        # numpy list rows (y) first, and cols second
        x = np.random.randint(self.im.shape[1] - side - 1)
        y = np.random.randint(self.im.shape[0] - side - 1)
        return (x, y, side)
        
    def __generate_car_coords_OLD(self, car_coords, x, y, side, rot_angle=None):
        patch = self.im[y:y+side, x:x+side, :]
        if rot_angle != None:
            patch = rotate_bound(patch, rot_angle)
        #patch = cv2.cvtColor(patch, cv2.COLOR_BGR2RGB)
        coords = []
        for i in range(len(car_coords[0])):
            x = car_coords[1][i]
            y = car_coords[0][i]
            x1 = max(0, x - 24)
            y1 = max(0, y - 24)
            x2 = min(patch.shape[1] - 1, x + 24)
            y2 = min(patch.shape[0] - 1, y + 24)
            #cv2.rectangle(patch, (x1, y1), (x2, y2), (255, 0, 0), 1)
            #print(x1, y1, x2, y2)
            #append cattegory - always 0 in this case x1 y1 x2 y2
            coords.append((0, x1, y1, x2, y2))
        return (patch, coords)
    
    def get_random_car_patch(self, min_cars=1, max_cars=None, rotate=True, flip_flag=True):
        patch = np.zeros((0,0,0))
        while len(patch) == 0:
            (x, y, side) = self.__create_random_x_y()      
            rot_angle = np.random.randint(359)
            annotated_patch = self.im_annotated_cars[y:y+side, x:x+side, 2]
            car_coords = annotated_patch.nonzero()
            ncars = len(car_coords[0])
            
            
            
            if ncars > 0:
                print(ncars)
            
            
            
            if flip_flag:
                if x % 2 == 0:
                    annotated_patch = cv2.flip(annotated_patch, 0)
                    car_coords = annotated_patch.nonzero()
            if rotate:
                annotated_patch = rotate_bound(annotated_patch, rot_angle)                
                car_coords = annotated_patch.nonzero()
            #if there are cars annotated in the patch
            if ncars >= min_cars and (max_cars == None or ncars <= max_cars):
                if rotate:
                    #when we rotate a pixel from the first image, transforms into 1 to 4 pixels in the new one
                    #also they can range in +/-2 pixels 
                    coords_0 = [car_coords[0][0]]
                    coords_1 = [car_coords[1][0]] 
                    for i in range(1, len(car_coords[0])):
                        if (abs(car_coords[0][i] - car_coords[0][i-1]) > 4 
                                or abs(car_coords[1][i] - car_coords[1][i-1]) > 4):
                            coords_0.append(car_coords[0][i])
                            coords_1.append(car_coords[1][i])
                    car_coords = np.array([coords_0, coords_1])
                    (patch, coords) = self.__generate_car_coords(car_coords, x, y, side, rot_angle, flip_flag)
                else:
                    (patch, coords) = self.__generate_car_coords(car_coords, x, y, side, flip_flag=flip_flag)
        return(patch, coords)
        
    
    def __create_random_x_y(self):
        #select a patch side within 96 (2 cars) to 256 (6 cars) side
        side = self.__min_side + np.random.randint(self.__max_side - self.__min_side)
        #select random x, y starting point
        # numpy list rows (y) first, and cols second
        x = np.random.randint(self.im.shape[1] - side - 1)
        y = np.random.randint(self.im.shape[0] - side - 1)
        return (x, y, side)
        
    def __generate_car_coords(self, car_coords, x, y, side, rot_angle=None, flip_flag=False):
        patch = self.im[y:y+side, x:x+side, :]
        if flip_flag:
            if x % 2 == 0:
                patch = cv2.flip(patch, 0)
        if rot_angle != None:
            patch = rotate_bound(patch, rot_angle)
        #patch = cv2.cvtColor(patch, cv2.COLOR_BGR2RGB)
        coords = []
        for i in range(len(car_coords[0])):
            x = car_coords[1][i]
            y = car_coords[0][i]
            x1 = max(0, x - 24)
            y1 = max(0, y - 24)
            x2 = min(patch.shape[1] - 1, x + 24)
            y2 = min(patch.shape[0] - 1, y + 24)
            cv2.rectangle(patch, (x1, y1), (x2, y2), (255, 0, 0), 1)
            #print(x1, y1, x2, y2)
            #append cattegory - always 0 in this case x1 y1 x2 y2
            coords.append((0, x1, y1, x2, y2))
            
            
            
        print('============', len(coords))
        
        
        
        return (patch, coords)
        
    
        
    
```

```python
#from https://github.com/Guanghan/darknet/blob/master/scripts/convert.py

"""
Created on Wed Dec  9 14:55:43 2015
This script is to convert the txt annotation files to appropriate format needed by YOLO 
@author: Guanghan Ning
Email: gnxr9@mail.missouri.edu
"""

import os
from os import walk, getcwd
from PIL import Image

classes = ["car"]

def convert(size, box):
    #box is (x1,x2,y1,y2)
    dw = 1./size[0]
    dh = 1./size[1]
    #center in pixels
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    #width / height in pixels
    w = box[1] - box[0]
    h = box[3] - box[2]
    #convert to relative
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)
    
    
"""-------------------------------------------------------------------""" 

""" Configure Paths"""   
data_filename = "test"


mypath = "/home/georgib/projects/cowc_detect/data/" + data_filename + "/"
outpath = "/home/georgib/projects/cowc_detect/data/" + data_filename + "/labels"

cls = "car"
if cls not in classes:
    exit(0)
cls_id = classes.index(cls)

wd = getcwd()
list_file = open('%s/%s_%s_list.txt'%(wd, cls, data_filename), 'w')

""" Get input text file list """
txt_name_list = []
temp_list = []
for (dirpath, dirnames, filenames) in walk(mypath + '/labels_not_yolo'):
    temp_list.extend(filenames)
    break
#clear the jpegs
for f in temp_list:
    #if f[-4:] == '.txt':
    txt_name_list.append(f)
print(txt_name_list)

""" Process """
for txt_name in txt_name_list:
    # txt_file =  open("Labels/stop_sign/001.txt", "r")

    """ Open input text files """
    txt_path = mypath + '/labels_not_yolo/' + txt_name
    print("Input:" + txt_path)
    txt_file = open(txt_path, "r")
    lines = []
    for l in txt_file:
        print('line', l)
        lines.append(l)
    #lines = txt_file.read().split('\r\n')   #for ubuntu, use "\r\n" instead of "\n"
    print('lines', lines)
    """ Open output text files """
    txt_outpath = outpath + '/' + txt_name
    print("Output:" + txt_outpath)
    txt_outfile = open(txt_outpath, "w")
    
    
    """ Convert the data to YOLO format """
    ct = 0
    for line in lines:
        #print('lenth of line is: ')
        #print(len(line))
        #print('inside for')
        #print('\n')
        if(len(line) >= 2):
            ct = ct + 1
            print(line + "\n")
            elems = line.split(' ')
            print(elems)
            #we have the class id as first elem so our xmin = elems[1] and xmax=elems[3]
            xmin = elems[1]
            xmax = elems[3]
            ymin = elems[2]
            ymax = elems[4]
            #
            #img_path = str('%s/images/%s/%s.JPEG'%(wd, cls, os.path.splitext(txt_name)[0]))
            img_path = str('%s%s/%s.jpeg'%(mypath, 'images', os.path.splitext(txt_name)[0]))
            print('img_path', img_path)
            #t = magic.from_file(img_path)
            #wh= re.search('(\d+) x (\d+)', t).groups()
            im=Image.open(img_path)
            w= int(im.size[0])
            h= int(im.size[1])
            #w = int(xmax) - int(xmin)
            #h = int(ymax) - int(ymin)
            # print(xmin)
            print(w, h)
            b = (float(xmin), float(xmax), float(ymin), float(ymax))
            bb = convert((w,h), b)
            print(bb)
            txt_outfile.write(str(cls_id) + " " + " ".join([str(a) for a in bb]) + '\n')

    """ Save those images with bb into list"""
    if(ct != 0):
        list_file.write('%s%s/%s.jpeg\n'%(mypath, 'images', os.path.splitext(txt_name)[0]))
        #list_file.write('%s/images/%s/%s.JPEG\n'%(wd, cls, os.path.splitext(txt_name)[0]))
                
list_file.close() 

#Checked for inconsistencies for numpy, opencv and yolo (normal images)
```
